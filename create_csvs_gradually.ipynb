```python
import pandas
import time
from twitter_account_enrichment import *

twitter_human_bots_dataset_path = "twitter_human_bots_dataset/twitter_human_bots_dataset.csv"
```

```python
twitter_human_bots_data = pandas.read_csv(twitter_human_bots_dataset_path, index_col=0)
accounts_ids = twitter_human_bots_data.id.astype(str).tolist()
```

# Followers Enrichment

```python
followers_per_account = get_accounts_followers(accounts_ids[:])
```

```python
followers_counter = 0
followers_nones_counter = 0
for account_id, followers in followers_per_account.items():
    if followers is not None:
        followers_counter += len(followers)
    else:
        followers_nones_counter += 1
len(followers_per_account), followers_counter, followers_nones_counter
```

```python
columns_names = ["followed_id", "follower_id", "follower_username", "follower_name", "follower_created_at", 
                 "follower_description", "follower_location", "follower_pinned_tweet_id", 
                 "follower_followers_count", "follower_following_count", "follower_tweet_count", 
                 "follower_listed_count"]
new_data = {column_name: [] for column_name in columns_names}
          
for followed_id, followers in followers_per_account.items():
    if followers is None:
        for column_name in columns_names:
            if column_name == "followed_id":
                new_data[column_name].append(followed_id)
            else:
                new_data[column_name].append(None)
        continue
    for follower in followers:
        for column_name in columns_names:
            if column_name == "followed_id":
                new_data[column_name].append(followed_id)
            elif column_name in ("follower_followers_count", "follower_following_count", "follower_tweet_count", 
                               "follower_listed_count"):
                new_data[column_name].append(follower["public_metrics"][column_name[9:]])
            else:
                new_data[column_name].append(follower.get(column_name[9:]))
```

```python
df = pd.DataFrame(data=new_data)
df
```

```python
df = pd.DataFrame(data=new_data)
df.to_csv(f"twitter_human_bots_dataset/followers{int(time.time())}.csv", index=False)
```

# Tweets Enrichment

```python
tweets_per_account = get_accounts_tweets([370890435])
```

```python
tweets_counter = 0
tweets_nones_counter = 0
for account_id, tweets in tweets_per_account.items():
    if tweets is not None:
        tweets_counter += len(tweets)
    else:
        tweets_nones_counter += 1
len(tweets_per_account), tweets_counter, tweets_nones_counter
```

```python
tweets_per_account[list(tweets_per_account.keys())[0]]
```

```python
columns_names = ["id","attachments","context_annotations","conversation_id","created_at","account_id", 
                 "in_reply_to_user_id","lang","retweet_count","reply_count","like_count","quote_count",
                 "reply_settings","source","text", "mentions", "hashtags", "urls"]
new_data = {column_name: [] for column_name in columns_names}

for account_id, tweets in tweets_per_account.items():
    if tweets is None:
        for column_name in columns_names:
            if column_name == "account_id":
                new_data[column_name].append(str(account_id))
            else:
                new_data[column_name].append(None)
        continue
    for tweet in tweets:
        for column_name in columns_names:
            if column_name in ("retweet_count", "reply_count", "like_count", "quote_count"):
                new_data[column_name].append(tweet["public_metrics"][column_name])
            elif column_name == "mentions":
                mentions_usernames = []
                if "mentions" in tweet.get("entities", []):
                    mentions_usernames = [m["username"] for m in tweet["entities"]["mentions"]]
                new_data[column_name].append(mentions_usernames)
            elif column_name == "hashtags":
                hashtags_tags = []
                if "hashtags" in tweet.get("entities", []):
                    hashtags_tags = [m["tag"] for m in tweet["entities"]["hashtags"]]
                new_data[column_name].append(hashtags_tags)
            elif column_name == "urls":
                urls_expanded_url = []
                if "urls" in tweet.get("entities", []):
                    urls_expanded_url = [m["expanded_url"] for m in tweet["entities"]["urls"]]
                new_data[column_name].append(urls_expanded_url)
            elif column_name == "account_id":
                new_data[column_name].append(str(account_id))
            elif column_name in ("id", "conversation_id"):
                new_data[column_name].append(str(tweet.get(column_name)))
            else:
                new_data[column_name].append(tweet.get(column_name))
```

```python
df = pd.DataFrame(data=new_data)
df
```

```python
df = pd.DataFrame(data=new_data)
df.to_csv(f"twitter_human_bots_dataset/tweets{int(time.time())}.csv", index=False)
```
