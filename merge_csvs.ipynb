```python
import pandas
import time
from twitter_account_enrichment import *

tweets_per_account = get_accounts_tweets(nontweetsids[1500:])
```

```python
tweets_counter = 0
tweets_nones_counter = 0
for account_id, tweets in tweets_per_account.items():
    if tweets is not None:
        tweets_counter += len(tweets)
    else:
        tweets_nones_counter += 1
len(tweets_per_account), tweets_counter, tweets_nones_counter
```

```python
columns_names = ["id","attachments","context_annotations","conversation_id","created_at","account_id", 
                 "in_reply_to_user_id","lang","retweet_count","reply_count","like_count","quote_count",
                 "reply_settings","source","text", "mentions", "hashtags", "urls"]
new_data = {column_name: [] for column_name in columns_names}

for account_id, tweets in tweets_per_account.items():
    if tweets is None:
        for column_name in columns_names:
            if column_name == "account_id":
                new_data[column_name].append(str(account_id))
            else:
                new_data[column_name].append(None)
        continue
    for tweet in tweets:
        for column_name in columns_names:
            if column_name in ("retweet_count", "reply_count", "like_count", "quote_count"):
                new_data[column_name].append(tweet["public_metrics"][column_name])
            elif column_name == "mentions":
                mentions_usernames = []
                if "mentions" in tweet.get("entities", []):
                    mentions_usernames = [m["username"] for m in tweet["entities"]["mentions"]]
                new_data[column_name].append(mentions_usernames)
            elif column_name == "hashtags":
                hashtags_tags = []
                if "hashtags" in tweet.get("entities", []):
                    hashtags_tags = [m["tag"] for m in tweet["entities"]["hashtags"]]
                new_data[column_name].append(hashtags_tags)
            elif column_name == "urls":
                urls_expanded_url = []
                if "urls" in tweet.get("entities", []):
                    urls_expanded_url = [m["expanded_url"] for m in tweet["entities"]["urls"]]
                new_data[column_name].append(urls_expanded_url)
            elif column_name == "account_id":
                new_data[column_name].append(str(account_id))
            else:
                new_data[column_name].append(tweet.get(column_name))
```

```python
df = pd.DataFrame(data=new_data)
df.to_csv(f"twitter_human_bots_dataset/tweets{int(time.time())}.csv", index=False)
```
