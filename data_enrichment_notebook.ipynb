```python
import pandas
import twitter_account_enrichment as twitter_utils

twitter_human_bots_dataset_path = "twitter_human_bots_dataset.csv"
```

```python
twitter_human_bots_data = pandas.read_csv(twitter_human_bots_dataset_path)
```

```python
res = []
users_lookup_params = {
    "user.fields": "id,name,username,created_at,description,location,pinned_tweet_id,protected,public_metrics,"
                   "verified,withheld"
}

for i in range(0, twitter_human_bots_data.count().id, 100):
    current_account_ids = twitter_human_bots_data.id.iloc[i:i+100].astype(str).tolist()
    users_lookup_url = twitter_utils.get_users_lookup_url(current_account_ids)
    json_response = twitter_utils.connect_to_endpoint(users_lookup_url, users_lookup_params)
    res.extend(json_response['data'])
    
    
```

```python
len(res)
```

```python
accounts_data = {}

for r in res:
    accounts_data[r['id']] = r
```

```python
len(accounts_data)
```

```python
list(accounts_data.values())[10]
```

```python
columns_names = ["id", "name", "username", "following_count", "tweet_count", "listed_count", "created_at", 
                 "pinned_tweet_id", "protected", "verified", "description", "withheld", "location", "account_type"]

new_data = {column_name: [] for column_name in columns_names}

for row in accounts_data.values():
    current_account_id = int(row['id'])
    for column_name in columns_names:
        if column_name == "account_type":
            try:
                new_data[column_name].append(twitter_human_bots_data.loc[
                    twitter_human_bots_data['id'] == current_account_id].account_type.iloc[0])
            except Exception as e:
                print(type(current_account_id))
                print(current_account_id)
                raise e
        elif column_name in ("following_count", "tweet_count", "listed_count"):
            new_data[column_name].append(row["public_metrics"][column_name])
        else:
            new_data[column_name].append(row.get(column_name))
```

```python
columns_names1 = ["id", "username", "following_count", "tweet_count", "listed_count", "created_at", 
                 "pinned_tweet_id", "protected", "verified", "withheld", "location", "account_type"]

new_data1 = {column_name: [] for column_name in columns_names1}

for row in accounts_data.values():
    current_account_id = int(row['id'])
    for column_name1 in columns_names1:
        if column_name1 == "account_type":
            new_data1[column_name1].append(twitter_human_bots_data.loc[
                twitter_human_bots_data['id'] == current_account_id].account_type.iloc[0])
        elif column_name1 in ("following_count", "tweet_count", "listed_count"):
            new_data1[column_name1].append(row["public_metrics"][column_name1])
        else:
            new_data1[column_name1].append(row.get(column_name1))
```

```python
df = pandas.DataFrame(data=new_data)
df1 = pandas.DataFrame(data=new_data1)
df.to_csv("enriched_twitter_human_bots_dataset_second.csv", index=False)
df1.to_csv("enriched_twitter_human_bots_dataset_second1.csv", index=False)
```
